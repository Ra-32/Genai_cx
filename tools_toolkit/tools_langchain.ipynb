# custom tools


```python
# step 1 create a function
def multiply(a,b):
  """---Multiply of two number---"""
  return a*b
```

```python
# step 2 add hint

def multiply(a:int,b:int)->int:
  """---Multiply of two number---"""
  return a*b
```

```python
# add tool decorator
from langchain_core.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    """Multiplies two numbers together"""
    return a * b
```

```python
result=multiply.invoke({'a':2,'b':3})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

# method 2 structeredtool

```python
from pydantic import BaseModel, Field
from langchain_core.tools import tool,StructuredTool
```

```python
from typing import Required
class MultiplySchema(BaseModel):
  a:int=Field(required=True,description='first number')
  b:int=Field(Required=True,description='second number')
```

```python
def multily_fun(a:int,b:int) -> int:
  """Multiply two number"""
  return a*b
```

```python
mutliply_tool=StructuredTool.from_function(
    func=multily_fun,
    name='multiply',
    description='Multiply two number',
    args_schema=MultiplySchema
)
```

```python
result=mutliply_tool.invoke({'a':6,'b':3})
print(result)
```

```python
print(mutliply_tool.name)
print(mutliply_tool.description)
print(mutliply_tool.args)
```

# basetool

```python
from langchain.tools import BaseTool
from typing import Optional,Type
```

```python
class MultiplySchema(BaseModel):
  a:int=Field(required=True,description='first number')
  b:int=Field(Required=True,description='second number')
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplySchema

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_base=MultiplyTool()
result=multiply_base.invoke({'a':2,'b':3})
print(result)
```

```python
print(multiply_base.name)
print(multiply_base.description)
print(multiply_base.args_schema)
```

# toolkit collection of all tool

```python
from langchain_core.tools import tool

@tool
def add(a:int,b:int)->int:
  """add of two number"""
  return a+b

@tool
def subtract(a:int,b:int)->int:
  """subtract of two number"""
  return a-b

```

```python
class Mathtoolkit:
  def get_tool(self):
    return [add,subtract]
```

```python
toolkit=Mathtoolkit()
```

```python
tools=toolkit.get_tool()
for tool in tools:
  print(tool.name ,'=>', tool.description)

```
